<h1> Some checks on distrib convergence </h1>

```python
n = 4
delta_alpha = 0.01
delta_tau = 0.01

alpha_tab = np.linspace(-7, 13, int(20 / delta_alpha))
tau_tab = np.linspace(0,100,int(100/delta_tau))

integrand = np.exp(-1j * alpha_tab[:,np.newaxis] * tau_tab[np.newaxis,:]) * np.ones(len(alpha_tab))[:,np.newaxis] * tau_tab[np.newaxis,:]**n
distrib = delta_tau * np.sum(integrand, axis=1)
```

```python
fig, ax = plt.subplots(1,2,figsize=(7,3), dpi=resol)

ax[0].plot(alpha_tab, np.real(distrib))
ax[1].plot(alpha_tab, np.imag(distrib))

for i in range(2):
    ax[i].grid()
    ax[i].set_axisbelow(True) 

plt.tight_layout()
plt.show()
```

```python
distrib_integration = delta_alpha * np.sum(distrib)
real_part = np.real(distrib_integration)
imag_part = np.imag(distrib_integration)

print("Real part = ", real_part) 
print("Imaginary part = ", imag_part)
```
